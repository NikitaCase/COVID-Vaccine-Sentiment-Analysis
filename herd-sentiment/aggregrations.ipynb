```python
# For stats 
# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_rel.html
from scipy import stats
```

```python
import pandas as pd 
```

```python
df = pd.read_csv('csv/tweets.csv')
```

```python
df.head()
```

```python
def GetManufacturer(txt):
    txt = txt.lower()
    if 'moderna' in txt:
        return 'mo'
    elif 'pfizer' in txt: 
        return 'pf'
    elif 'astra' in txt:
        return 'az'
    else:
        return '0'
```

```python
df['manufacturer'] = df['text'].apply(GetManufacturer)
df.tail()
```

```python
df = df[['id_str','verified','retweet_count','tweet_favourite_count','Subjectivity','Polarity','Sentiment','manufacturer']]
```

```python
df['Subjectivity'] = df['Subjectivity'].round(decimals=1)
df['Polarity'] = df['Polarity'].round(decimals=1)
df
```

```python
df2 = df.drop(columns=['id_str'])
df2
```

```python
df3 = df2.groupby(by ='manufacturer').sum()
df3
```

```python
df4 = df2.groupby(by =['manufacturer','Polarity']).sum()
df4
```

```python
df_new = df2[['manufacturer','Polarity']]
```

```python
by_man = df_new.groupby(by=['manufacturer'])
```

```python
pf_df2 = by_man['Polarity'].value_counts(normalize=True).loc['pf']
az_df2 = by_man['Polarity'].value_counts(normalize=True).loc['az']
mo_df2 = by_man['Polarity'].value_counts(normalize=True).loc['mo']
un_df2 = by_man['Polarity'].value_counts(normalize=True).loc['0']
```

```python
polarity = pd.concat(
    [pf_df2,az_df2,mo_df2,un_df2], 
    axis=1, 
    keys=['Pfizer-BioNTech','AstraZeneca','Moderna','Unknown']
)
polarity
```

```python
polarity = polarity.reset_index()
polarity
```

```python
polarity.to_csv('polarity.csv',index=False, encoding='UTF-8')
```
